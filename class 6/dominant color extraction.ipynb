```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("flower.jpg" )
img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
print(img.shape)
```

```python
plt.imshow(img)
plt.show()
```

```python
all_pixel = img.reshape((-1,3))
all_pixel.shape
```

```python
from sklearn.cluster import KMeans
```

```python
dominant_color = 4
km = KMeans(n_clusters=dominant_color)
```

```python
km.fit(all_pixel)
```

```python
centers = km.cluster_centers_
print(centers)
```

```python
centers = np.array(centers, dtype = "uint8")
print(centers)
```

```python
km.labels_
```

```python
np.unique(km.labels_,return_counts = True)
```

```python
new_image = np.zeros((img.shape[0] * img.shape[1],3) , dtype = "uint8")
print(new_image.shape)
```

```python
for i in range(new_image.shape[0]):
    new_image[i] = centers[km.labels_[i]]
```

```python
new_image = new_image.reshape((img.shape[0] , img.shape[1],3))
```
