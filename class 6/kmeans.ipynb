```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X , y = make_blobs(n_samples=900, n_features=2 , centers=5,random_state=3)
```

```python
k = 5 
color = ["green" ,  "red" , "blue" , "yellow" ,  "orange"]

np.random.seed(3)
clusters = {}
for i in range(k):
    center = 10*(np.random.random((2,))-0.5)
#     print(center)
    point = []
    cluster = {
        'center':center,
        'points' : point,
        'color' : color[i]
    }
    clusters[i] = cluster
    
print(clusters)
```

```python
def distance(x1,x2):
    return np.sqrt(sum((x1-x2)**2))
```

```python
def assignPointsToCluster(clusters):
    for i in range(X.shape[0]):
        dist =[]
        curr = X[i]
        
        for kx in range(k):
            d = distance(curr,clusters[kx]["center"] )
            dist.append(d)
        current_cluster = np.argmin(dist)
        clusters[current_cluster]["points"].append(curr)
```

```python
def updateClusters(clusters):
    for kx in range(k):
        pts = np.array(clusters[kx]["points"])
        
        if pts.shape[0]>0:
            new_u = pts.mean(axis = 0)
            clusters[kx]["center"] = new_u
            clusters[kx]["points"]  = []
```

```python
def plotClusters(clusters):
    for kx in range(k):
        pts = np.array(clusters[kx]["points"])
        plt.scatter(pts[:,0] , pts[:,1] , color = clusters[kx]['color'])

        uk = clusters[kx]["center"]
        print(uk)
        plt.scatter(uk[0] ,uk[1] , color = "black" , marker ="+")
```

```python

assignPointsToCluster(clusters)
plotClusters(clusters)   
updateClusters(clusters)
```
