```python
def hypothesis(x,theta):
    y_ = 0.0
    n = x.shape[0]
    for i in range(n):
        y_  += (theta[i]*x[i])
    return y_


def error(X,Y, theta):
    m = X.shape[0]
    total_error = 0
    
    for i in range(m):
        y_ = hypothesis(X[i],  theta)
        total_error+=((y_-Y[i])**2)
    return total_error/m
```

```python

def gradient(X,y,theta):
    m,n = X.shape
    
    grad = np.zeros((n,))
    for j in range(n):
        for i in range(m):
            y_ = hypothesis(X[i],theta)
            grad[j] += (y_ - y[i])*X[i][j]

    return grad/m
```

```python
def gradientDescent(X,Y, lr = 0.1,max_step=300):
    m,n = X.shape
    theta = np.zeros((n,))
    error_list = []
    theta_list = []
    
    for i in range(max_step):
        grad = gradient(X,Y , theta)
        err = error(X,Y ,theta)
        
#         theta = theta -lr*grad
        for j in range(n):
            theta[j] =  theta[j] - lr*grad[j] 
        error_list.append(err)
        theta_list.append(theta)
        
    return error_list , theta_list , theta
```
