```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(5)
x2 = x**2
x3 = x**3
print(x,x2,x3)
```

```python
plt.plot(x)
plt.plot(x2)
plt.plot(x3)
plt.show()
```

```python
x = np.array([1,2,3,4,5,6])
y = np.array([10,20,30,40,50,60])
```

```python
plt.scatter(x,y ,color="red",marker="^")
plt.title("stock data")
plt.xlabel("Days")
plt.ylabel("Price")

plt.show()
```

```python
numbers = np.random.randint(0,10,5)
indices = np.arange(5)

numbers2 = np.random.randint(0,10,5)
indices2 = np.arange(5)


print(numbers)
print(numbers2)
```

```python
plt.bar(indices,numbers,0.25,label = "apple")
plt.bar(indices2+0.25,numbers2,0.25, label="banana")
plt.bar(indices2+0.5, numbers,0.25 , label = "mango")
plt.legend()
plt.show()
```
